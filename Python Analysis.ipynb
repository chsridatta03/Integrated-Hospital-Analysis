# EDA

```python
appointments_df['appointment_date'] = pd.to_datetime(appointments_df['appointment_date'])

appointments_df['appointment_hour'] = appointments_df['appointment_time'].apply(lambda t: t.hour)
```

```python
daily_volume = appointments_df.groupby('appointment_date').size()
```

```python
daily_volume.head(10)
```

Hourly Volume

```python
hourly_volume = appointments_df.groupby('appointment_hour').size()
```

```python
hourly_volume.head(10)
```

```python
plt.figure(figsize=(10,4))

daily_hour = appointments_df.groupby('appointment_hour').size()
ax = daily_hour.plot(kind='bar',color='#b8860b')

plt.title('Appointments by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Appointments')

ax.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.6)

for i, value in enumerate(daily_hour.values):
    ax.text(i, value + 0.5, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
pivot = appointments_df.pivot_table(index='day_of_week',columns='appointment_hour',values='appointment_id',aggfunc='count').fillna(0)

plt.figure(figsize=(12,4))
plt.imshow(pivot, aspect='auto', cmap='Reds') 

plt.yticks(range(len(pivot.index)), pivot.index)
plt.xlabel('Hour of Day')
plt.title('Appointments heatmap (day vs hour)')
plt.colorbar()

plt.tight_layout()
plt.show()
```

No show and Cancellation Diagnostics

```python

status_counts = appointments_df['status'].value_counts(normalize=False).rename('count')
no_show_by_doc = appointments_df[appointments_df['status']=='No-show'].groupby('doctor_id').size().sort_values(ascending=False)
no_show_by_day = appointments_df[appointments_df['status']=='No-show'].groupby('day_of_week').size().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']).fillna(0)

plt.figure(figsize=(8,3))
no_show_by_day.plot(kind='bar')
plt.title('No-shows by Day of Week')
plt.xlabel('Day')
plt.ylabel('No-show Count')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()


ns_patients = appointments_df[appointments_df['status']=='No-show'].groupby('patient_id').size().sort_values(ascending=False)
ns_patients.head()
```

```python

billings_df['bill_date'] = pd.to_datetime(billings_df['bill_date'])
monthly_revenue = billings_df.groupby(billings_df['bill_date'].dt.to_period('M'))['amount'].sum()

plt.figure(figsize=(9,3))
monthly_revenue.plot(marker='o')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.tight_layout()
plt.show()
```

```python
rev_by_method = billings_df.groupby('payment_method')['amount'].sum().sort_values(ascending=False)
rev_by_method.head()
```

```python

patients_df['date_of_birth'] = pd.to_datetime(patients_df['date_of_birth'])
patients_df['age'] = (pd.Timestamp.today() - patients_df['date_of_birth']).dt.days // 365

plt.figure(figsize=(8, 3))

plt.hist(
    patients_df['age'].dropna(),
    bins=25,
    edgecolor='black'
)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

patients_df['age_group'] = pd.cut(patients_df['age'],
                                  bins=[0,20,35,50,65,90],
                                  labels=['0-20','21-35','36-50','51-65','66+'])

visits = appointments_df.groupby('patient_id').size().rename('visit_count')
demo = patients_df.merge(visits, on='patient_id', how='left')

g = sns.FacetGrid(demo, col='gender', height=4)
g.map_dataframe(sns.boxplot, x='age_group', y='visit_count')
plt.show()
```
